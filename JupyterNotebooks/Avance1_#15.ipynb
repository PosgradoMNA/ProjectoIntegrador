```python
import requests
import json
import time
import pandas as pd
import yfinance as yf

from datetime import date, timedelta
from google.colab import userdata
```

```python
df_SP500 = pd.read_csv('S&P500_List.csv')
```

```python
SP500_symbol = df_SP500['Symbol'].to_list()
SP500_name = df_SP500['Security'].to_list()
```

```python
daily_register_stocks = {
    'Symbol': [],
    'Name': [],
    'Open': [],
    'High': [],
    'Low': [],
    'Close': [],
    'Status_Stock': [],
}
```

```python
end_date = date.fromisoformat('2024-04-26')
# end_date = date.today()

start_date = end_date - timedelta(days=1)
```

```python
for i in range(len(SP500_symbol)):
  stock = SP500_symbol[i]
  name = SP500_name[i]

  daily_register_stocks['Symbol'].append(stock)
  daily_register_stocks['Name'].append(name)

  try:
    stock_data = yf.download(stock, start_date, end_date)

    daily_register_stocks['Open'].append(stock_data['Open'].values[0])
    daily_register_stocks['High'].append(stock_data['High'].values[0])
    daily_register_stocks['Low'].append(stock_data['Low'].values[0])
    daily_register_stocks['Close'].append(stock_data['Close'].values[0])
    daily_register_stocks['Status_Stock'].append('OK')
    pass

  except Exception as e:
    daily_register_stocks['Open'].append(' ')
    daily_register_stocks['High'].append(' ')
    daily_register_stocks['Low'].append(' ')
    daily_register_stocks['Close'].append(' ')
    daily_register_stocks['Status_Stock'].append(f"{e}")
    pass
```

```python
# print(f"CSV: {len(SP500_symbol)}\n")

error = 0
for key in daily_register_stocks.keys():
  if len(daily_register_stocks[key]) != len(SP500_symbol):
    print(f"MISSMATCH IN: {key}")
  # print(f"{key}: {len(daily_register_stocks[key])}")

if error == 0:
  print('Count OK')
```
