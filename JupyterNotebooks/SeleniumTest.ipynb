```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
allData = soup.find_all("article",{"class":"IFHyqb DeXSAc"})
```

```python
for data in allData[:5]:
    link = data.find('a').get('href')
    link = f"https://news.google.com{link[1:]}"
    title = data.find('a', {'class':'JtKRv'}).text
    author = data.find('span', {'class':'PJK1m'}).text
    date = data.find('time', {'class':'hvbAAd'}).text

    article = Article(link)

    try:
        article.download()
        article.parse()
        
        text = article.text
    except Exception as error:
        text = F"ERROR: {error}"
    
    print(link)
    print(title)
    print(author)
    print(date)
    print(text)
    print("")
```
